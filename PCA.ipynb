## seperating the features and the labels

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

Apply Standard Scaling


```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

Find Covariance Matrix

```python
covariance_matrix = np.cov(x)
print('Covariance Matrix : \n', covariance_matrix)
```

finding eigen vectors and eigen values

```python
eigen_values, eigen_vectors = np.linalg.eig(covariance_matrix)
```

```python
print('Eigen vectors :\n', eigen_vectors)
```

```python
print('Eigen values :\n', eigen_values)
```

### selecting the principal components
to do this it is better to create the eigen pairs between the vectors and the values, this would help in not dropping too much information to create lower dimensional subspace. The lowest eigenvalues can be dropped as they consist of less information present in them.


```python
#making a list of eigen values and vectors
eigen_pairs = [(np.abs(eigen_values[i]), eigen_vectors[:, i]) for i in range(len(eigen_values))]
```

```python
#sorting the tuples from high to low
eigen_pairs.sort(key=lambda x: x[0], reverse=True)
```

```python
#confirming the list is correctly sorted by decreasing the eigenvalues
print('Eigen pair count :', len(eigen_pairs))
print('Eigenvalues in descending order :')
for i in eigen_pairs:
  print(i[0])
```

we need to decide the number of princiapal components to be chosen, for that it is better to use expalined variance that will tell us the amount of information that can be generated from each of the principal components.

```python
total = sum(eigen_values)
variance_explained = [(i/total)*100 for i in sorted(eigen_values, reverse=True)]
```

```python
with plt.style.context('dark_background'):
    plt.figure(figsize=(6, 4))
    plt.bar(range(178), variance_explained, alpha=0.5, align='center',
            label='individual explained variance')

    plt.ylabel('Explained variance ratio')
    plt.xlabel('Principal components')
    plt.legend(loc='best')
    plt.tight_layout()
```
